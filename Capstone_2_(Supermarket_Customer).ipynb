##Melihat Outlier pada setiap variable Numerik

```python
numerik = df.select_dtypes(include = 'number')

# ganti ukuran pigura (wadah grafik nya)
plt.figure(figsize=(10,30))

posisi = 1 # menginisiasi posisi mulai dari 1
for setiap_kolom in numerik.columns: # numerik.columns = nama-nama kolom dalam list
    plt.subplot(13,2,posisi)
    sns.boxplot(df[setiap_kolom])
    posisi = posisi + 1
    # agar tidak bertumpuk
    plt.tight_layout()
```

Berdasarkan Boxplot untuk setiap variable numerik dalam data set, dapat kita lihat pada kolom Year_Birth, Income, MntWines, MntFruits, MntMeatProducts, MntFishProducts, MntSweetProducts, MntGoldProds, NumDealsPurchases, NumWebPurchases, NumCatalogPurchases, NumWebVisitsMonth memiliki outlier. Untuk AcceptedCmp1-5, Complain dan Respons sangat wajar terdapat outlier karena value dikolom tersebut hanya terdiri dari 0 dan 1.

##Menghitung Jumlah Outlier menggunakan Inter Quartile Range (IQR) * 1.5

```python
list_jumlah_outliers = []
for setiap_kolom in numerik.columns:
    q3 = numerik.describe()[setiap_kolom]['75%']
    q1 = numerik.describe()[setiap_kolom]['25%']
    iqr = q3 - q1
    whis = 1.5 * iqr
    upper_bound = q3 + whis
    lower_bound = q1 - whis

    jumlah_outliers = len(numerik[(numerik[setiap_kolom] > upper_bound) | (numerik[setiap_kolom] < lower_bound)])
    list_jumlah_outliers.append(jumlah_outliers)

pd.DataFrame({
                'kolom': numerik.columns,
                'jumlah outliers': list_jumlah_outliers

})
```

```python
#presentase outlier
pd.DataFrame({
                'kolom': numerik.columns,
                'presentase outliers': [jumlah/len(df) for jumlah in list_jumlah_outliers]
})
```

Pada presentase Jumlah Outlier dari setiap kolom, dapat kita lihat yang memiliki outlier diatas 10% yaitu pada kolom MntFruits,MntSweetProducts, NumCatalogPurchases maka dari itu nanti kita akan menentukan berdasarkan tujuan penelitian dan business knowledge dalam menangani outlier tersebut.

###Rata-rata Pengeluaran Per Pembelian:
Apakah rata-rata pengeluaran per pembelian berbeda antara pelanggan yang membeli melalui Web, Catalog, dan Store?

```python
# Hitung total pengeluaran untuk setiap kanal.
total_web = df_copy['NumWebPurchases'] * (df_copy['MntWines'] + df_copy['MntFruits'] + df_copy['MntMeatProducts'] + df_copy['MntFishProducts'] + df_copy['MntSweetProducts'] + df_copy['MntGoldProds'])
total_catalog = df_copy['NumCatalogPurchases'] * (df_copy['MntWines'] + df_copy['MntFruits'] + df_copy['MntMeatProducts'] + df_copy['MntFishProducts'] + df_copy['MntSweetProducts'] + df_copy['MntGoldProds'])
total_store = df_copy['NumStorePurchases']* (df_copy['MntWines'] + df_copy['MntFruits'] + df_copy['MntMeatProducts'] + df_copy['MntFishProducts'] + df_copy['MntSweetProducts'] + df_copy['MntGoldProds'])

# Hitung rata-rata pengeluaran per pembelian.
avg_web = total_web.sum() / df_copy['NumWebPurchases'].sum()
avg_catalog = total_catalog.sum() / df_copy['NumCatalogPurchases'].sum()
avg_store = total_store.sum() / df_copy['NumStorePurchases'].sum()

print("Rata-rata pengeluaran per pembelian di Web:", avg_web)
print("Rata-rata pengeluaran per pembelian di Catalog:", avg_catalog)
print("Rata-rata pengeluaran per pembelian di Store:", avg_store)
```

```python
#Bandingkan rata-rata pengeluaran
import matplotlib.pyplot as plt

# Data untuk Barplot
kanal = ['Web', 'Catalog', 'Store']
rata_rata = [avg_web, avg_catalog, avg_store]

# Membuat Barplot
colors= ['blue', 'orange', 'green']
plt.bar(kanal, rata_rata, color=colors)
plt.xlabel('Kanal Pembelian')
plt.ylabel('Rata-rata Pengeluaran per Pembelian')
plt.title('Perbandingan Rata-rata Pengeluaran per Pembelian pada Kanal')
plt.show()
```

Pelanggan cenderung mengeluarkan uang lebih banyak ketika berbelanja melalui kanal catalog dibandingkan dengan kanal lainnya. Rata-rata pengeluaran per pembelian melalui catalog jauh lebih tinggi dibandingkan dengan web dan store.

Kanal web memiliki rata-rata pengeluaran per pembelian yang lebih rendah dibandingkan dengan catalog, tetapi lebih tinggi dibandingkan dengan store. Hal ini mengindikasikan bahwa pelanggan mungkin lebih sering membeli barang dengan harga yang lebih terjangkau melalui web.

Kanal store memiliki rata-rata pengeluaran per pembelian yang paling rendah di antara ketiga kanal. Ini bisa jadi karena beberapa faktor, seperti adanya batasan pilihan produk, pengaruh lingkungan sekitar, atau adanya promo khusus di toko.

Kemungkinan alasan mengapa pelanggan cenderung mengeluarkan uang lebih banyak saat berbelanja melalui catalog:
  - Efek katalog: Katalog seringkali menampilkan produk dengan kualitas tinggi dan desain yang menarik, sehingga dapat memicu keinginan konsumen untuk membeli produk yang lebih mahal.
  - Kurangnya perbandingan harga: Ketika berbelanja melalui catalog, konsumen mungkin tidak memiliki banyak pilihan untuk membandingkan harga dengan produk serupa di tempat lain.
  - Pengalaman belanja yang lebih personal: Katalog seringkali memberikan informasi produk yang lebih detail dan personal, sehingga konsumen merasa lebih yakin untuk melakukan pembelian.

Untuk menetapkan target realistis dalam meningkatkan Frekuensi Pembelian, Rata-rata Pengeluaran, Pengaruh Sosial (Complaints), dan Efektivitas Kampanye Pemasaran dalam konteks supermarket, berikut ini adalah rekomendasinya:

1. Frekuensi Pembelian

- Target: Tingkatkan tingkat pembelian ulang (repeat purchase rate) setidaknya 5-10% per tahun.

- Strategi: Terapkan strategi keterlibatan pasca-pembelian yang dipersonalisasi. Ini bisa mencakup pengiriman pengingat yang ditargetkan berdasarkan pola pembelian pelanggan, seperti saran produk terkait. Selain itu, menggunakan otomatisasi pemasaran untuk memicu tindakan secara real-time dapat secara signifikan meningkatkan retensi pelanggan.

2. Rata-rata Pengeluaran per Pembelian

- Target: Tetapkan tujuan untuk meningkatkan ukuran keranjang rata-rata sebesar 10-15%.

- Strategi: Strategi cross-sell dan up-sell sangat efektif. Misalnya, memanfaatkan email transaksional untuk merekomendasikan produk pelengkap atau menawarkan paket bundling dan promosi dapat mendorong pengeluaran yang lebih tinggi per transaksi. Rekomendasi yang digerakkan oleh AI berdasarkan minat eksternal pelanggan juga dapat memperluas pengeluaran mereka di berbagai kategori produk.

3. Pengaruh Sosial (Complaints)

- Target: Kurangi penurunan pembelian terkait keluhan sebesar 20%.

- Strategi: Kembangkan pendekatan proaktif terhadap manajemen keluhan. Ini mencakup respon yang cepat terhadap keluhan dan menggunakan data dari keluhan untuk meningkatkan produk dan layanan. Pelanggan yang puas, terutama mereka yang masalahnya diselesaikan dengan efisien, lebih mungkin untuk terus berbelanja di Supermarket.

4. Efektivitas Kampanye Pemasaran

- Target: Tingkatkan ROI kampanye sebesar 15-20% dan tingkatkan tingkat konversi dengan fokus pada audiens yang ditargetkan.

- Strategi: Untuk mengukur dan meningkatkan efektivitas kampanye, tetapkan tujuan dan metrik yang jelas seperti ROI, tingkat konversi, dan biaya akuisisi pelanggan. Menggunakan kerangka pengukuran kampanye yang terstruktur dengan baik dapat membantu mengidentifikasi tren dan membuat penyesuaian berbasis data untuk mengoptimalkan kampanye yang sedang berjalan.
